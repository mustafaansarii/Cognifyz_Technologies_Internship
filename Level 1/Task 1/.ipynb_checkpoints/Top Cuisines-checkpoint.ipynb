```python
Cuisines=df['Cuisines'].value_counts()
Cuisines.head(3)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=Cuisines.head(3).index, y=Cuisines.head(3).values, palette='viridis')
plt.title('Top 3 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
```

```python
cuisines_split = df['Cuisines'].str.split(',').explode()

# Use value_counts() to get the frequency of each cuisine
cuisine_counts = cuisines_split.value_counts()

# Get the top 3 cuisines
top_3_cuisines = cuisine_counts.head(3).index

# Filter the DataFrame to include only rows with the top 3 cuisines
df_top_3_cuisines = df[df['Cuisines'].str.split(',').explode().isin(top_3_cuisines)]

# Calculate the percentage for each cuisine among the top 3
total_restaurants_top_3_cuisines = len(df_top_3_cuisines)
cuisine_percentages_top_3 = (df_top_3_cuisines['Cuisines'].str.split(',').explode().value_counts(normalize=True) * 100)

# Display the percentage of restaurants serving each of the top 3 cuisines
print("Percentage of Restaurants Serving Each Top 3 Cuisine:")
print(cuisine_percentages_top_3)
```
